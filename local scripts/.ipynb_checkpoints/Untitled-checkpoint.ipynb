```python
# Species, Taxonomy, MutSpec(12 col), Codon usage of 12 (64 col), Codon usage of ND6 (64 col)
# LIGHT chain
Suppl_table = pd.DataFrame(columns=['Species', 'Taxonomy', 
                                    'A>T', 'A>G', 'A>C', 'T>A', 'T>G', 'T>C', 'G>A', 'G>T', 'G>C', 'C>A', 'C>T', 'C>G', 
                                    'AAA', 'AAC', 'AAG', 'AAT', 'ACA', 'ACC', 'ACG', 'ACT', 'AGA', 'AGC', 'AGG', 'AGT', 'ATA', 'ATC',
                                    'ATG', 'ATT', 'CAA', 'CAC', 'CAG', 'CAT', 'CCA', 'CCC', 'CCG', 'CCT', 'CGA', 'CGC', 'CGG', 'CGT',
                                    'CTA', 'CTC', 'CTG', 'CTT', 'GAA', 'GAC', 'GAG', 'GAT', 'GCA', 'GCC', 'GCG', 'GCT', 'GGA', 'GGC', 
                                    'GGG', 'GGT', 'GTA', 'GTC', 'GTG', 'GTT', 'TAA', 'TAC', 'TAG', 'TAT', 'TCA', 'TCC', 'TCG', 'TCT', 
                                    'TGA', 'TGC', 'TGG', 'TGT', 'TTA', 'TTC', 'TTG', 'TTT', 
                                    'AAA_ND6', 'AAC_ND6', 'AAG_ND6', 'AAT_ND6', 'ACA_ND6', 'ACC_ND6', 'ACG_ND6', 'ACT_ND6', 'AGA_ND6', 
                                    'AGC_ND6', 'AGG_ND6', 'AGT_ND6', 'ATA_ND6', 'ATC_ND6', 'ATG_ND6', 'ATT_ND6', 'CAA_ND6', 'CAC_ND6', 
                                    'CAG_ND6', 'CAT_ND6', 'CCA_ND6', 'CCC_ND6', 'CCG_ND6', 'CCT_ND6', 'CGA_ND6', 'CGC_ND6', 'CGG_ND6', 
                                    'CGT_ND6', 'CTA_ND6', 'CTC_ND6', 'CTG_ND6', 'CTT_ND6', 'GAA_ND6', 'GAC_ND6', 'GAG_ND6', 'GAT_ND6', 
                                    'GCA_ND6', 'GCC_ND6', 'GCG_ND6', 'GCT_ND6', 'GGA_ND6', 'GGC_ND6', 'GGG_ND6', 'GGT_ND6', 'GTA_ND6', 
                                    'GTC_ND6', 'GTG_ND6', 'GTT_ND6', 'TAA_ND6', 'TAC_ND6', 'TAG_ND6', 'TAT_ND6', 'TCA_ND6', 'TCC_ND6', 
                                    'TCG_ND6', 'TCT_ND6', 'TGA_ND6', 'TGC_ND6', 'TGG_ND6', 'TGT_ND6', 'TTA_ND6', 'TTC_ND6', 'TTG_ND6', 
                                    'TTT_ND6'])

# species = 'Mus_musculus'
total_species = np.union1d(CU_ND6_df.Species, ms.Species)
for species in total_species:
    if species in ms.Species.values:
        mutspec = ms[ms.Species == species].values[0][1:] # array 
    else:
        mutspec = np.full(12, pd.NA)
    if species in CU_12_df.Species.values:
        taxonomy = CU_ND6_df[CU_ND6_df.Species == species].Taxonomy.iloc[0]
        CU_12 = CU_12_df[CU_12_df.Species == species].iloc[:, 1:-1].values[0]
        if 'ND6' in CU_ND6_df[CU_ND6_df.Species == species].Gene.values:
            CU_ND6 = CU_ND6_df[(CU_ND6_df.Species == species) & (CU_ND6_df.Gene == 'ND6')].iloc[:, 2:-6].values[0]
        else: CU_ND6 = np.full(64, pd.NA)
    else:
        taxonomy = pd.NA
        CU_12 = np.full(64, pd.NA)
        CU_ND6 = np.full(64, pd.NA)
    
    # if species in gl.Scientific_name.values:
    #     genlength = float(gl[gl.Scientific_name == species]['GenerationLength_d'].values[0])
    # else: 
    #     genlength = pd.NA
    # if species in ft.Species.values:
    #     fishtemp = float(ft[ft.Species == species].groupby('Species').mean().iloc[0].values[0])
    # else: 
    #     fishtemp = pd.NA
    # finaldeltarho = float(deltaRho_ShiftMean[species][224])
    row = np.concatenate((species, taxonomy, mutspec, CU_12, CU_ND6), axis=None) # + [genlength] + [fishtemp] + [finaldeltarho]
    Suppl_table.loc[len(Suppl_table)] = row
Suppl_table.head()
```
